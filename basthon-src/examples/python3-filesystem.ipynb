```python
import os

# mon_image.png n'est pas présent
os.listdir()
```

```python
import math
from PIL import Image, ImageDraw

w, h = 200, 200
bbox = [(10, 10), (w - 10, h - 10)]


img = Image.new("RGB", (w, h), "white")
dctx = ImageDraw.Draw(img)
dctx.pieslice(bbox, start=30, end=330, fill="#ffc938", outline="#366f9e", width=5)
del dctx
img = img.resize((w // 2, h // 2), resample=Image.Resampling.LANCZOS)
img.show()

img.save("mon_image.png")
```

```python
# maintenant elle est là !
os.listdir()
```

```python
from p5 import *

def setup():
    global img
    img = loadImage("filesystem:/mon_image.png")
    createCanvas(300, 300)

def draw():
    image(img, 0, 0);

run()
```

```python
from p5 import *

i = 0
img = None

def preload():
    global img
    img = loadImage("filesystem:/mon_image.png")

def setup():
    createCanvas(300, 300)

def draw():
    global i
    background(250)
    image(img, i % 300, (i // 300) * 60 % 300);
    i += 1


run(preload=preload)
```
